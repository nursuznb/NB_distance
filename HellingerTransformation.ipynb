```python
import pandas as pd
import numpy as np

df = pd.read_csv("marine_benthic.csv")
```

```python
abund_cols = df.select_dtypes(include=[np.number]).columns
df_abund = df[abund_cols]
```

```python
# Row-wise relative abundance
row_sums = df_abund.sum(axis=1)

df_rel = df_abund.div(row_sums, axis=0)

# Square-root transform
df_hell = np.sqrt(df_rel)
```

```python
df_hellinger_full = pd.concat(
    [df.drop(columns=abund_cols), df_hell],
    axis=1
)
```

```python
df_hellinger_full.to_csv(
    "benthic_hellinger.csv",
    index=False
)
```
