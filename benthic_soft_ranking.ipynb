```python
#real data
#post-Benthic-Hellinger complete

import pandas as pd

df = pd.DataFrame({
    "Silhouette": [0.205,	0.254,	0.261,	0.313,	0.238],
    "DB": [1.237,	1.384,	1.479,	1.406,	1.412],
    "CH": [12.9,	7.96,	6.30,	6.13,	9.20],
    "Dunn": [0.275,	0.377,	0.362,	0.214,	0.489]
})

# Normalize: Larger good
def norm_high(x):
    return (x - x.min())/(x.max()-x.min())

# Normalize: Smaller good
def norm_low(x):
    return (x.max()-x)/(x.max()-x.min())

df["Sil_score"] = norm_high(df["Silhouette"])
df["CH_score"]  = norm_high(df["CH"])
df["DB_score"]  = norm_low(df["DB"])
df["Dunn_score"]= norm_high(df["Dunn"])

# Final soft score (average)
df["SoftScore"] = df[["Sil_score","CH_score","DB_score","Dunn_score"]].mean(axis=1)

print(df["SoftScore"])
```
